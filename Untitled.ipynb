```python
import os, re
import numpy as np
import matplotlib.pyplot as plt
import geoip2.database
import uuid
import pandas as pd
raw_data = {}
ipv6Count = np.int16(0)
ipv4Count = np.int16(0)
def clean_str(s): return re.sub(r'[\x00-\x08\x0A-\x1F]', '', s)
for root, dirs, files in os.walk('netDb'):
    for file in files:
        full_path = os.path.join(root, file)
        with open(full_path, "rb") as f:
            content = f.read()
            try:
                text = content.decode('utf-8', errors='ignore')
                pattern = re.compile(r'([\w\x00-\x1F]+)=([\s\S]+?);', re.DOTALL)
                #  re.findall(r'\w+?=.+;', text)
                matches = pattern.findall(text)
                uid = str(uuid.uuid4())
                raw_data[uid] = {}
                for raw_key, raw_value in matches:
                    key = clean_str(raw_key)
                    value = clean_str(raw_value)
                    raw_data[uid][key] = value
            except Exception as e:
                print(f"Error reading {full_path}: {e}")



df = pd.DataFrame([raw_data])
row_dicts = df.iloc[0]
raw_data = {}
del raw_data

hosts = [d.get('host') for d in row_dicts if isinstance(d, dict) and 'host' in d]
hosts = set(hosts)
for host in hosts:
    if re.match(r'^([\da-fA-F]{1,4}:){7}[\da-fA-F]{1,4}$', host) is not None:
        ipv6Count += 1
    else:
        ipv4Count +=1

labels = ['IPv4', 'IPv6']
sizes = [ipv4Count, ipv6Count]
colors = ['#66b3ff', '#ff9999']
explode = (0.1, 0)  

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('IPv4 vs IPv6')
plt.axis('equal') 
plt.show()

reader = geoip2.database.Reader('GeoLite2-City.mmdb')    
```

```python
from collections import Counter
country_counter = Counter()
region_counter = Counter()
city_counter = Counter()
for ip in hosts:
    try:
        response = reader.city(ip)
        country = response.country.name or "Unknown"
        region = response.subdivisions.most_specific.name or "Unknown"
        city = response.city.name or "Unknown"
        
        country_counter[country] += 1
        region_counter[region] += 1
        city_counter[city] += 1

    except geoip2.errors.AddressNotFoundError:
        country_counter["Unknown"] += 1
        region_counter["Unknown"] += 1
        city_counter["Unknown"] += 1
    except ValueError:
        pass
        
```

```python
countries = list(country_counter.keys())
country_counts = list(country_counter.values())

countries_np = np.array(countries)
country_counts_np = np.array(country_counts, dtype=np.int32)

regions = list(region_counter.keys())
region_counts = list(region_counter.values())

regions_np = np.array(regions)
region_counts_np = np.array(region_counts, dtype=np.int32)

cities = list(city_counter.keys())
city_counts = list(city_counter.values())

cities_np = np.array(cities)
city_counts_np = np.array(city_counts, dtype=np.int32)  

countries = []
regions = []
cities = []
del countries
del regions
del cities
```

```python
plt.figure(figsize=(12, 6))
plt.bar(countries_np, country_counts_np)
plt.xticks(rotation=90)  
plt.xlabel('Countries')
plt.ylabel('Count IP')
plt.title('IP for Countries')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

regions = regions_np
counts = region_counts_np

def plot_top_regions(top_n=20):
    indices = np.argsort(counts)[-top_n:]
    top_regions = regions[indices]
    top_counts = counts[indices]

    plt.figure(figsize=(10, max(5, top_n*0.3)))
    plt.barh(top_regions, top_counts, color='skyblue')
    plt.xlabel('Count IP')
    plt.title(f'Top {top_n} regions by IP count')
    plt.tight_layout()
    plt.show()
plot_top_regions(top_n=100)
```

```python
versions = [d.get('version') for d in row_dicts if isinstance(d, dict) and 'version' in d]
count_versions = Counter(versions)
labels = list(count_versions.keys())
sizes = list(count_versions.values())


plt.figure(figsize=(10, 6))
plt.bar(labels, sizes, color='skyblue')
plt.xlabel('Version')
plt.ylabel('Count')
plt.title('Version by count')
plt.xticks(rotation=45) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
data = []

for d in row_dicts:
    if isinstance(d, dict):
        ip = d.get('host')
        version = d.get('version')
        port = d.get('port')  # Extract port too
        if ip and version:
            try:
                response = reader.city(ip)
                country = response.country.name or "Unknown"
                if country != "Unknown":
                    latitude = response.location.latitude
                    longitude = response.location.longitude
                    data.append({
                        'ip': ip,
                        'version': version,
                        'country': country,
                        'port': port,
                        'lat': latitude,
                        'lon': longitude
                    })

            except Exception:
                pass

df = pd.DataFrame(data)
del data
top_countries = df['country'].value_counts().head(10).index.tolist()

country_version_counts = df.groupby(['country', 'version']).size().reset_index(name='count')

filtered = country_version_counts[country_version_counts['country'].isin(top_countries)]

filtered['country_idx'] = filtered['country'].astype('category').cat.codes
filtered['version_idx'] = filtered['version'].astype('category').cat.codes

plt.figure(figsize=(14,8))
scatter = plt.scatter(
    filtered['country_idx'], filtered['version_idx'],
    s=filtered['count'] * 10,  # Point size proportional to count
    c=filtered['count'],        # Color by count
    cmap='viridis', alpha=0.7, edgecolors='w'
)

plt.colorbar(scatter, label='Router Count')
plt.xticks(ticks=range(len(top_countries)), labels=top_countries, rotation=45, ha='right')
plt.yticks(ticks=range(len(filtered['version'].unique())), labels=filtered['version'].astype('category').cat.categories)

plt.xlabel('Country')
plt.ylabel('Router Version')
plt.title('Distribution of Router Versions by Country (Top 10 Countries)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


port_counts = df['port'].value_counts().head(20)

plt.figure(figsize=(12,6))
port_counts.plot(kind='bar')
plt.xlabel('Port')
plt.ylabel('Count')
plt.title('Top 20 Most Common Ports')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
