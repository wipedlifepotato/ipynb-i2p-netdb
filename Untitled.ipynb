```python
import os, re
import numpy as np
import matplotlib.pyplot as plt
import geoip2.database
IPsv4 = []
IPsv6 = []
for root, dirs, files in os.walk('netDb'):
    for file in files:
        full_path = os.path.join(root, file)
        with open(full_path, "rb") as f:
            content = f.read()
            try:
                text = content.decode('utf-8', errors='ignore')
                ipv4 = re.findall(r"\d+\.\d+\.\d+\.\d+", text)
                ipv6 = re.findall(r'(?i)\b(?:[0-9a-f]{1,4}:){7}[0-9a-f]{1,4}\b', text)
                if ipv4:
                    #print(f"In file {full_path} found IPs: {ipv4}")
                    IPsv4 += ipv4
                    IPsv6 += ipv6
            except Exception as e:
                print(f"Error reading {full_path}: {e}")
IPsv4 = set(IPsv4)
IPsv6 = set(IPsv6)
print(f'IPv4 count: {len(IPsv4)}')
print(f'IPv6 count: {len(IPsv6)}')
reader = geoip2.database.Reader('GeoLite2-City.mmdb')    
```

```python
from collections import Counter
country_counter = Counter()
region_counter = Counter()
city_counter = Counter()
for ip in IPsv4:
    try:
        response = reader.city(ip)
        country = response.country.name or "Unknown"
        region = response.subdivisions.most_specific.name or "Unknown"
        city = response.city.name or "Unknown"
        
        country_counter[country] += 1
        region_counter[region] += 1
        city_counter[city] += 1

    except geoip2.errors.AddressNotFoundError:
        country_counter["Unknown"] += 1
        region_counter["Unknown"] += 1
        city_counter["Unknown"] += 1
        
```

```python
countries = list(country_counter.keys())
country_counts = list(country_counter.values())

countries_np = np.array(countries)
country_counts_np = np.array(country_counts, dtype=np.int32)

regions = list(region_counter.keys())
region_counts = list(region_counter.values())

regions_np = np.array(regions)
region_counts_np = np.array(region_counts, dtype=np.int32)

cities = list(city_counter.keys())
city_counts = list(city_counter.values())

cities_np = np.array(cities)
city_counts_np = np.array(city_counts, dtype=np.int32)  
```

```python
plt.figure(figsize=(12, 6))
plt.bar(countries_np, country_counts_np)
plt.xticks(rotation=90)  
plt.xlabel('Countries')
plt.ylabel('Count IP')
plt.title('IP for Countries')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

regions = regions_np
counts = region_counts_np

def plot_top_regions(top_n=20):
    indices = np.argsort(counts)[-top_n:]
    top_regions = regions[indices]
    top_counts = counts[indices]

    plt.figure(figsize=(10, max(5, top_n*0.3)))
    plt.barh(top_regions, top_counts, color='skyblue')
    plt.xlabel('Count IP')
    plt.title(f'Top {top_n} regions by IP count')
    plt.tight_layout()
    plt.show()
plot_top_regions(top_n=100)
```
